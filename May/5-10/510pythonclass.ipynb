```python
class UserInfo(object):
    users_list = {}

    def __init__(self, user_id, user_pw, user_name, user_email):
        self.user_id = user_id
        self.user_pw = user_pw
        self.user_name = user_name
        self.user_email = user_email
        self.users_list[user_name] = self

    def log_in(self, id, pw):
        if id == self.users_list[self.user_name].user_id:
            if pw == self.users_list[self.user_name].user_pw:
                print(f'{self.user_name}님 정상적으로 로그인 되었습니다.\n반갑습니다.')
            else:
                print('비밀번호를 확인해 주세요')
        else:
            print('아이디를 확인해 주세요')
    
    def email_send(self):
        print(f"{self.user_name}님 이메일이 {self.user_email}로 정상적으로 발송되었습니다.\n감사합니다.")

a = UserInfo('abb', 'aaa', 'latacha', 'sdf@sdf.sdf')
a.log_in('abb', 'aaa')
a.email_send()
```
